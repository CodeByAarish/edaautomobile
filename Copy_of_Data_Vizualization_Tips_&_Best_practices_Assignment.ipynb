```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data = df, hue = 'Species')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length v/s Petal Width")
plt.show
```

```python
# Create a bar chart showing the average sepal length for each species of iris flowers
average_sepal_length = df.groupby('Species')['SepalLengthCm'].mean()
average_sepal_length
```

```python
# Create a box plot of sepal width for each species of iris flowers
sns.boxplot(x='Species', y='SepalWidthCm', data = df)
plt.xlabel("Different Species")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Width (cm) for each species!")
plt.show()
```

```python
# Write your code here
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(data=np.c_[X, y], columns=np.append(feature_names, 'diagnosis'))

# Rename the diagnosis labels for better visualization
df['diagnosis'] = df['diagnosis'].replace({0: 'Benign', 1: 'Malignant'})
```

```python
# Set a visually appealing color palette
colors = sns.color_palette('pastel')

# Distribution of mean radius values for malignant and benign diagnoses
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='mean radius', hue='diagnosis', fill=True, palette=colors)

# Add labels and title
plt.title('Distribution of Mean Radius Values for Malignant and Benign Diagnoses')
plt.xlabel('Mean Radius')
plt.ylabel('Density')

# Display the chart
plt.show()
```

```python
# Heatmap of correlations between different features
plt.figure(figsize=(18, 8))
corr_matrix = df.iloc[:, :30].corr()
sns.heatmap(data=corr_matrix, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Breast Cancer Features')

# Display the chart
plt.show()
```

```python
# Bar chart of distribution of mean texture values for malignant and benign diagnoses
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='diagnosis', y='mean texture', palette=colors)

# Add labels and title
plt.title('Distribution of Mean Texture Values for Malignant and Benign Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Mean Texture')

# Display the chart
plt.show()
```

```python
# Scatter plot matrix to explore relationships between features
sns.set(style='ticks', color_codes=True)
diag_colors = {'Benign': colors[0], 'Malignant': colors[1]}
sns.pairplot(data=df, vars=['mean radius', 'mean smoothness', 'worst concavity'], hue='diagnosis',
             palette=diag_colors, markers=['o', 'D'], plot_kws={'alpha': 0.6})

# Display the scatter plot matrix
plt.show()
```

```python
# Box plot comparing worst texture values between malignant and benign diagnoses
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='diagnosis', y='worst texture', palette=colors)

# Add labels and title
plt.title('Comparison of Worst Texture Values between Malignant and Benign Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Worst Texture')

# Display the box plot
plt.show()
```

```python
# Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the Wine Quality dataset
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
white_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'

# Choose either red or white wine dataset
wine_df = pd.read_csv(red_wine_url, sep=';')
```

```python
# Scatter plot of alcohol content vs. quality rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=wine_df, x='alcohol', y='quality', hue='quality', palette='viridis')

# Add labels and title
plt.title('Scatter Plot of Alcohol Content vs. Quality Rating')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality Rating')
plt.legend(title='Quality', loc='upper left')

# Display the scatter plot
plt.show()
```

```python
# Line chart of pH trend over time for a specific type of wine
plt.figure(figsize=(10, 6))
sns.lineplot(data=wine_df, x='free sulfur dioxide', y='pH', color='mediumorchid')

# Add labels and title
plt.title('pH Trend over Time for Wine Quality')
plt.xlabel('Free Sulfur Dioxide')
plt.ylabel('pH')

# Display the line chart
plt.show()
```

```python
# Heatmap of correlations between different wine properties
plt.figure(figsize=(10, 8))
corr_matrix = wine_df.corr()
sns.heatmap(data=corr_matrix, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Wine Properties')

# Display the heatmap
plt.show()
```

```python
# Bar chart of average acidity for each type of wine
wine_type_avg_acidity = wine_df.groupby('type')['fixed acidity'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(data=wine_type_avg_acidity.reset_index(), x='type', y='fixed acidity', palette='RdPu')

# Add labels and title
plt.title('Average Acidity for Each Wine Type')
plt.xlabel('Wine Type')
plt.ylabel('Average Acidity')

# Display the bar chart
plt.show()
```

```python
# Pie chart showing distribution of wine quality ratings
quality_counts = wine_df['quality'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Pastel1'))

# Add title
plt.title('Distribution of Wine Quality Ratings')

# Display the pie chart
plt.show()
```

2. Choose the Right Type of Visualization: Select appropriate types of visualizations, such as bar charts, pie charts, or stacked bar charts, based on the data you want to display and the insights you want to communicate.

3. Choose Specific Colors: Use colors that are visually appealing, distinguishable, and appropriate for the context of the data. Avoid using too many colors or colors that are too bright or too dull, as they can affect the readability and interpretability of the visualization.

4. Label Specifically: Ensure that all labels, including axis labels, titles, and legends, are clear, concise, and specific. Use descriptive labels that provide context and make it easy for the audience to understand the information being presented.

2. Create a pie chart showing the proportion of male and female passengers on the Titanic. Use a color palette that is visually appealing and ensure that the slices of the pie chart are labeled with the corresponding gender.

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
# Calculate the proportion of survivors and non-survivors for each passenger class
survivors_by_class = titanic_df.groupby(['class', 'survived'])['survived'].count().unstack()
survivors_by_class
total_by_class = titanic_df.groupby('class')['survived'].count()
total_by_class
proportion_by_class = survivors_by_class.div(total_by_class, axis=0)

# Set a visually appealing color palette
colors = sns.color_palette('pastel')

# Create the stacked bar chart
proportion_by_class.plot(kind='bar', stacked=True, color=colors)

# Add labels and title
plt.title('Proportion of Survivors and Non-Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Proportion')
plt.legend(['Non-Survivor', 'Survivor'], loc='upper right')

# Display the chart
plt.show()
```

```python
# Pie Chart showing the proportion of male and female passengers on the Titanic:

# Count the number of male and female passengers
gender_counts = titanic_df['sex'].value_counts()

# Set a visually appealing color palette
colors = sns.color_palette('pastel')

# Create the pie chart
plt.pie(gender_counts, labels=gender_counts.index, colors=colors, autopct='%1.1f%%')

# Add a title
plt.title('Proportion of Male and Female Passengers')

# Display the chart
plt.show()
```

```python
# Histogram showing the distribution of passenger ages:
# Create the histogram
sns.histplot(data=titanic_df, x='age', bins=10, kde=True, color=sns.color_palette('pastel')[0])

# Add labels and title
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the chart
plt.show()
```

```python
# Scatter Plot showing the relationship between passenger age and fare paid:

# Create the scatter plot
sns.scatterplot(data=titanic_df, x='age', y='fare', hue='survived', palette='Pastel1')

# Add labels and legend
plt.title('Age vs Fare paid')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(['Did not survive', 'Survived'], loc='upper right')

# Display the chart
plt.show()
```

```python
# Bar Chart showing the number of passengers in each age range who survived or did not survive:

# Create age ranges
age_ranges = pd.cut(titanic_df['age'], bins=[0, 18, 30, 50, 100])
titanic_df['age_range'] = age_ranges

# Count the number of passengers in each age range who survived or did not survive
count_by_age_range = titanic_df.groupby(['age_range', 'survived'])['survived'].count().unstack()

# Set a visually appealing color palette
colors = sns.color_palette('pastel')

# Create the bar chart
count_by_age_range.plot(kind='bar', stacked=True, color=colors)

# Add labels and title
plt.title('Number of Passengers in Each Age Range by Survival Status')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.legend(['Did not survive', 'Survived'], loc='upper right')

# Display the chart
plt.show()
```
